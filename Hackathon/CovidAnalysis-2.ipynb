# **Data Analysis and Visualization**

*imports for libraries we used*

```python
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
import numpy as np
import datetime
import pandas as pd
from datetime import datetime
from matplotlib import dates as mpl_dates
```

## **Reading The Data**

```python
Data = pd.read_csv(r"covidAgeGender.csv")
Data1 = pd.read_csv(r"covidAgeGender.csv")
Datagrp= pd.read_csv(r"covidAgeGender.csv")
Data.head(42)
```

## **Checking The Null Values**

```python

features = list(Data.columns)
features
```

```python
# showing the null values in data 
sns.heatmap(Data.isnull(),yticklabels=False,cbar=False,cmap='Blues')
cols = list(Data.columns)
print('The Nulls In Data Columns:')
for c in cols:
    print(c, ': ' , Data[c].isna().sum() ,' from :', Data[Data.columns[0]].count() )
```

## **Total Cases By Gender**

```python
Data = Data.replace(['זכר'],'1')
Data = Data.replace(['נקבה'],'0')
Data = Data.replace(['לא ידוע'],'2')

counts = Data.groupby('gender')['weekly_cases'].sum()
print(" Weekly cases by gender" + str(counts))
gender = ['Female','Male','Other']
plt.bar(gender, counts, color ='midnightblue',
        width = 0.4)

plt.xlabel("Gender")
plt.ylabel("Total Cases")
plt.title("Total Cases By Gender")
plt.show()
```

## **Total Tests By Gender**

```python
counts1 = Data.groupby('gender')['weekly_tests_num'].sum()
print("Weekly tests by gender" + str(counts1))
gender = ['Female','Male','Other']
plt.bar(gender, counts1, color ='midnightblue',
        width = 0.4)

plt.xlabel("Gender")
plt.ylabel("Total Tests")
plt.title("Total Tests By Gender")
plt.show()
```

### **Converting The Date to Date Type**

```python
Data["first_week_day"] = pd.to_datetime(Data["first_week_day"],dayfirst=True,format="%d/%m/%Y")
Data = Data.sort_values(by="first_week_day")
```

```python
Data['morbidity']=Data['weekly_cases']/Data['weekly_tests_num']
```

# **PLOTS**

```python
weekly_morbidity=Data.groupby('first_week_day')['morbidity'].mean()
plt.figure(figsize=[20,10])
week = Data['first_week_day'].unique()
plt.plot(week, weekly_morbidity, color ='midnightblue')
plt.xlabel("Weeks")
plt.ylabel("Morbidity")
plt.title("Covid-19 Morbidity [2020-2022]")
plt.gcf().autofmt_xdate()
plt.show()
```

```python

# weekly_morbidity=Data.groupby('first_week_day')['morbidity'].mean()
# plt.figure(figsize=[20,10])
# week = Data['first_week_day'].unique()
# plt.plot(week, weekly_morbidity, color ='midnightblue')
# plt.xlabel("Weeks")
# plt.ylabel("Morbidity")
# plt.title("Covid-19 Morbidity [2020-2022]")
# plt.gcf().autofmt_xdate()
# plt.show()

# Datagrp["first_week_day"] = pd.to_datetime(Datagrp["first_week_day"],dayfirst=True,format="%d/%m/%Y")
Datagrp['morbidity']=Datagrp['weekly_cases']/Datagrp['weekly_tests_num']
Datagrp.index = pd.to_datetime(Datagrp['first_week_day'],format="%d/%m/%Y")
d = Datagrp.groupby(by=[Datagrp.index.month, Datagrp.index.year])['morbidity'].mean()
df1=pd.DataFrame()
months=[]
years=[]
morbidity=[]
for i in range(len(d)):
  months.append(d.index[i][0])
  years.append(d.index[i][1])
  morbidity.append(d.iloc[i])
```

```python

df1['months']=months
df1['years']=years
df1['morbidity']=morbidity


a = {'months': 1, 'years': 2020, 'morbidity': 0}
b = {'months': 2, 'years': 2020, 'morbidity': 0}

c = {'months': 4, 'years': 2022, 'morbidity': 0}
d = {'months': 5, 'years': 2022, 'morbidity': 0}
e = {'months': 6, 'years': 2022, 'morbidity': 0}
f = {'months': 7, 'years': 2022, 'morbidity': 0}
g = {'months': 8, 'years': 2022, 'morbidity': 0}
h = {'months': 9, 'years': 2022, 'morbidity': 0}
i = {'months': 10, 'years': 2022, 'morbidity': 0}
j = {'months': 11, 'years': 2022, 'morbidity': 0}
k = {'months': 12, 'years': 2022, 'morbidity': 0}



df1 = df1.append(a ,ignore_index=True)
df1 = df1.append(b,ignore_index=True)
df1 = df1.append(c,ignore_index=True)
df1 = df1.append(d,ignore_index=True)
df1 = df1.append(e,ignore_index=True)
df1 = df1.append(f,ignore_index=True)
df1 = df1.append(g,ignore_index=True)
df1 = df1.append(h,ignore_index=True)
df1 = df1.append(i,ignore_index=True)
df1 = df1.append(j,ignore_index=True)
df1 = df1.append(k,ignore_index=True)

df1.sort_values(by=['months'], inplace=True)

print(df1)
plt.figure(figsize=[22,10])

plt.xlabel("months")
plt.ylabel("morbidity")
plt.title("Covid-19 Morbidity according to Years")
plt.gcf().autofmt_xdate()
months1 = ['January' , 'February' , 'March' , 'April' , 'May' , 'June' , 'July' , 'August' , 'September' , 'October' , 'November' , 'December']
# plt.plot(df1['months'].unique(), df1['morbidity'].loc[df1['years']==2020], color ='midnightblue', label = "2020")
# plt.plot(df1['months'].unique(), df1['morbidity'].loc[df1['years']==2021], color ='aqua' , label = "2021")
# plt.plot(df1['months'].unique(), df1['morbidity'].loc[df1['years']==2022], color ='red' , label = "2022")

plt.plot(months1, df1['morbidity'].loc[df1['years']==2020], color ='midnightblue', label = "2020")
plt.plot(months1, df1['morbidity'].loc[df1['years']==2021], color ='aqua' , label = "2021")
plt.plot(months1, df1['morbidity'].loc[df1['years']==2022], color ='red' , label = "2022")

plt.legend()
plt.show()
```

```python

counts_genderWeeklyCases = Data.groupby(['first_week_day','gender'])['weekly_cases'].sum()
df=pd.DataFrame()

week=[]
gender=[]
cases=[]
for i in range(315):
  week.append(counts_genderWeeklyCases.index[i][0])
  gender.append(int(counts_genderWeeklyCases.index[i][1]))
  cases.append(counts_genderWeeklyCases[i])

df['week']=week
df['gender']=gender
df['cases']=cases

plt.figure(figsize=[20,10])

plt.xlabel("Months")
plt.ylabel("Number Of Total Cases")
plt.title("Total Cases in three years")
plt.gcf().autofmt_xdate()

plt.plot(df['week'].unique(), df['cases'].loc[df['gender']==0], color ='midnightblue', label = "Female")
plt.plot(df['week'].unique(), df['cases'].loc[df['gender']==1], color ='aqua' , label = "male")
plt.plot(df['week'].unique(), df['cases'].loc[df['gender']==2], color ='red' , label = "Other")


plt.legend()
plt.show()
```

```python
counts_genderWeeklyTests = Data.groupby(['first_week_day','gender'])['weekly_tests_num'].sum()
df=pd.DataFrame()

week=[]
gender=[]
tests=[]
for i in range(315):
  week.append(counts_genderWeeklyTests.index[i][0])
  gender.append(int(counts_genderWeeklyTests.index[i][1]))
  tests.append(counts_genderWeeklyTests[i])

df['week']=week
df['gender']=gender
df['Tests']=tests

plt.figure(figsize=[20,10])

plt.xlabel("Months")
plt.ylabel("Number Of Total Tests")
plt.title("Total Tests in three years")
plt.gcf().autofmt_xdate()

plt.plot(df['week'].unique(), df['Tests'].loc[df['gender']==0], color ='midnightblue', label = "Female")
plt.plot(df['week'].unique(), df['Tests'].loc[df['gender']==1], color ='aqua' , label = "male")
plt.plot(df['week'].unique(), df['Tests'].loc[df['gender']==2], color ='red' , label = "Other")



plt.legend()
plt.show()
```
